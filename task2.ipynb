#importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
population_data = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_2163544.csv', skiprows=4)
```

```python
year = '2022'
```

```python
population_data_year = population_data[['Country Name', year]]
```

```python
population_data_year.dropna(inplace=True)
```

```python
population_data_year.sort_values(by=year, ascending=False, inplace=True)
```

```python
top_countries = population_data_year.head(10)
```

```python
plt.figure(figsize=(15, 6))
plt.bar(top_countries['Country Name'], top_countries[year],edgecolor="Red",color ='maroon')
plt.title('Top 10 Countries by Population Count in {}'.format(year))
plt.xlabel('Country')
plt.ylabel('Population Count')
plt.xticks(rotation=45)
plt.show()
```
